```python
import pandas as pd
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
moviesdf = pd.read_csv("movies.csv")
moviesdf.head()
```

```python
ratingsdf = pd.read_csv("ratings.csv")
ratingsdf.head()
```

```python
moviesdf['year'] = moviesdf.title.str.extract('(\(\d\d\d\d\))',expand=False)
moviesdf['year'] = moviesdf.year.str.extract('(\d\d\d\d)',expand=False)
moviesdf['title'] = moviesdf.title.str.replace('(\(\d\d\d\d\))','')
moviesdf['title'] = moviesdf.title.apply(lambda x: x.strip())
moviesdf.head()
```

```python
moviesdf['genres']=moviesdf.genres.str.split('|')
moviesdf.head()
```

```python
moviesgenre = moviesdf.copy()
for index, row in moviesdf.iterrows():
    for genre in row['genres']:
        moviesgenre.at[index,genre]=1
moviesgenre = moviesgenre.fillna(0)
moviesgenre.head()
```

```python
ratingsdf = ratingsdf.drop('timestamp',1)
ratingsdf.head()
```

```python
userInput = [
            {'title':'Breakfast Club, The', 'rating':5},
            {'title':'Toy Story', 'rating':3.5},
            {'title':'Jumanji', 'rating':2},
            {'title':"Pulp Fiction", 'rating':5},
            {'title':'Akira', 'rating':4.5},
            {'title':'Yeh Jawaani Hai Deewani','rating':5}
         ] 
inputmovies = pd.DataFrame(userInput)
inputmovies
```

```python
inputId = moviesdf[moviesdf['title'].isin(inputmovies['title'].tolist())]
inputmovies = pd.merge(inputId,inputmovies)
inputmovies = inputmovies.drop('genres',1).drop('year',1)
inputmovies
```

```python
usermovies = moviesgenre[moviesgenre['movieId'].isin(inputmovies['movieId'].tolist())]
usermovies
```

```python
usermovies = usermovies.reset_index(drop=True)
usergenre = usermovies.drop('movieId',1).drop('title',1).drop('genres',1).drop('year',1)
usergenre
```

```python
inputmovies['rating']
```

```python
userprofile = usergenre.transpose().dot(inputmovies['rating'])
userprofile
```

```python
genretable = moviesgenre.set_index(moviesgenre['movieId'])
genretable = genretable.drop('movieId',1).drop('title',1).drop('genres',1).drop('year',1)
genretable
```

```python
recommendation = (genretable*userprofile).sum(axis=1)/(userprofile.sum())
recommendation
```

```python
recommendation = recommendation.sort_values(ascending=False)
recommendation.head(10)
```
